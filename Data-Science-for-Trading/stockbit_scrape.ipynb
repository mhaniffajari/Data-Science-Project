```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time
import random

# Function to simulate human typing
def human_typing(element, text):
    for char in text:
        element.send_keys(char)
        time.sleep(random.uniform(0.1, 0.3))

# Function to simulate random sleep
def human_sleep(min_sec=1, max_sec=3):
    time.sleep(random.uniform(min_sec, max_sec))

# Setup Chrome options
options = webdriver.ChromeOptions()
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36")
options.add_argument("--disable-blink-features=AutomationControlled")  # Helps avoid detection

driver = webdriver.Chrome(options=options)
browser = driver

# Open the login page
browser.get("https://stockbit.com/login")
human_sleep()

# Find and fill in the username and password fields
username_field = browser.find_element(By.ID, "username")
password_field = browser.find_element(By.ID, "password")

human_typing(username_field, "hanif123arsenal@gmail.com")
human_sleep()
human_typing(password_field, "Kimia2020")
human_sleep()

# Find and click the login button using ActionChains
login_button = browser.find_element(By.ID, "email-login-button")
ActionChains(browser).move_to_element(login_button).pause(random.uniform(0.5, 1.5)).click(login_button).perform()

# Optional manual CAPTCHA if still triggered
print("If CAPTCHA appears, please complete it.")
input("Press Enter after CAPTCHA is solved (if shown)...")

time.sleep(10)
wait = WebDriverWait(browser, 90)

# Wait for the financials page to load
browser.get("https://stockbit.com/symbol/INAF/financials")

dropdown = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '[data-cy="report-type"]')))
browser.execute_script("arguments[0].scrollIntoView();", dropdown)

# Click the dropdown
browser.execute_script('arguments[0].click();', dropdown)

# Select "Balance Sheet"
annual_option = browser.find_element(By.XPATH, '//option[text()="Balance Sheet"]')
annual_option.click()

# Select "Annual"
dropdown = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, '[data-cy="statement-type"]')))
dropdown.click()
annual_option = wait.until(EC.element_to_be_clickable((By.XPATH, '//option[text()="Annual"]')))
annual_option.click()

try:
    table = None

    try:
        table = wait.until(EC.presence_of_element_located((By.XPATH, '//table[@id="data_table_1"]')))
    except Exception as e:
        print("XPATH locator failed:", e)

    if table is None:
        try:
            table = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'fin-table')))
        except Exception as e:
            print("Class name locator failed:", e)

    if table is None:
        try:
            table = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'table#data_table_1')))
        except Exception as e:
            print("CSS selector locator failed:", e)

    if table is None:
        try:
            table = wait.until(EC.presence_of_element_located((By.ID, 'data_table_1')))
        except Exception as e:
            print("ID locator failed:", e)

    if table is None:
        try:
            table = wait.until(EC.presence_of_element_located((By.NAME, 'table_name')))
        except Exception as e:
            print("Name locator failed:", e)

    if table is None:
        try:
            table = wait.until(EC.presence_of_all_elements_located((By.TAG_NAME, 'table')))[0]
        except Exception as e:
            print("Tag name locator failed:", e)

    if table is None:
        raise Exception("Table not found using any locator")

    wait = WebDriverWait(browser, 240)
    table_html = table.get_attribute("outerHTML")

    with open("data_table_1.html", "w", encoding="utf-8") as file:
        file.write(table_html)

except Exception as e:
    print("Error:", e)

finally:
    browser.quit()
```

```python
from bs4 import BeautifulSoup
import pandas as pd

# Read the HTML file
with open("data_table_1.html", "r", encoding="utf-8") as html_file:
    html_content = html_file.read()

# Parse the HTML using BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Find the table element
table = soup.find("table")

# Initialize lists to store the header and data
header = []
data = []

# Extract table headers
header_row = table.find("tr")
for th in header_row.find_all("th"):
    header.append(th.text.strip())

# Extract table data
for row in table.find_all("tr")[1:]:  # Skip the header row
    row_data = [td.text.strip() for td in row.find_all("td")]
    data.append(row_data)

# Create a Pandas DataFrame
df = pd.DataFrame(data, columns=header)
```

```python
data = df.T
data.columns = data.iloc[0]
data = data.iloc[1: , :]
data = data.reset_index()
```

```python
data['tahun'] = data["index"].str.split().str[1]
```

```python
data = data.drop('index',axis=1)
```
