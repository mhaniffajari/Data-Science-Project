```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
client = pd.read_csv('client_data.csv')
price = pd.read_csv('price_data.csv')
```

```python
def preprocessing(df):
    """"
    Objective :
    Function for preprocessing data, known missing, duplicated values and basic stastics every column in dataframe and duplicated row
    df is dataframe
    """
    try:
        import pandas as pd
        variables = pd.DataFrame(columns=['Variable','Number of unique values','Percent of Null(%)','Number of Null','Type','Values'])
        for i, var in enumerate(df.columns):
            variables.loc[i] = [var, df[var].nunique(),df[var].isnull().sum()/df.shape[0]*100,df[var].isnull().sum(),df[var].dtypes,df[var].unique()]
        return (variables.set_index('Variable'))
    except:
        print('ada kesalahan penginputan')
```

```python
price['price_date'] = pd.to_datetime(price['price_date'])
```

```python
len(price['id'].unique())
```

```python
max_date_price = max(price['price_date'])
```

```python
price['days'] = price['price_date'].apply(lambda x:(max_date_price-x).days)
```

```python
price_unique = list(price['id'].unique())
```

```python
price_off_peak_sensitivity = []
for x in price_unique:
    from scipy.stats import linregress
    df = price[price['id']==x]
    c = linregress(df['price_off_peak_var'],df['days'])
    sloped = c.slope
    price_off_peak_sensitivity.append(sloped)
```

```python
price_sensitivity = pd.DataFrame({'id':price_unique,'price_off_peak_sensitivity'
                                 :price_off_peak_sensitivity})
```

```python
price = price.merge(price_sensitivity,on='id',how='left')
```
